# LLM Cybersecurity Benchmark Analysis
This notebook loads the pairwise comparison results and reproduces summary tables and charts.


```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
data_path = Path('..') / 'data' / 'prompt_results.csv'
df = pd.read_csv(data_path)
df.head()
```

```python
model_wins = df['winner'].value_counts().rename_axis('model').reset_index(name='wins')
vendor_wins = df['winner_vendor'].value_counts().rename_axis('vendor').reset_index(name='wins')
model_wins
```

```python
plt.figure(figsize=(10,5))
plt.bar(model_wins['model'], model_wins['wins'])
plt.title('Wins by Model')
plt.xlabel('Model')
plt.ylabel('Wins')
plt.xticks(rotation=35, ha='right')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.bar(vendor_wins['vendor'], vendor_wins['wins'])
plt.title('Wins by Vendor')
plt.xlabel('Vendor')
plt.ylabel('Wins')
plt.tight_layout()
plt.show()
```
